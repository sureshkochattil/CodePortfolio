```python
# Importing the required modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Reading the data 

raw_data = pd.read_csv("C:\\Users\\sures\\Downloads\\gapminder_new.csv")

# checking the file import

raw_data.head()

# We can see that there are missing values in the data
```

```python
# Converting the datatype from string to float
raw_data[['life', 'population']] = raw_data[['life', 'population']].apply(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
# Checking the conversion
print (raw_data.dtypes)
```

```python
# replacing null values of population between 1800 and 1950 with the mean population of the respective countries for the same period
filter_na_pop = raw_data[(raw_data['Year'] >= 1800) & (raw_data['Year'] <= 1950)]
replace_na_pop = filter_na_pop.groupby('Country')['population'].transform('mean')
raw_data['population'].fillna(replace_na_pop, inplace=True)

#replacing null values of population for years after 1950 with the mean of the population for that country
#replacing null values of income of countries with the mean income of all countries
raw_data['population'] = raw_data['population'].fillna(raw_data['population'].mean())
raw_data['income'] = raw_data['income'].fillna(raw_data['income'].mean())
```

```python
# Analysis 14: We will now do a simple regression model to arrive at the linear equation to predict life expectancy based
# on income

from sklearn import linear_model as lm
#from sklearn.linear_model import LinearRegression
#import statsmodels.api as sm

#creating a linear regression object
linear_reg = lm.LinearRegression()

income = raw_data[['income']]

life_expectancy = raw_data[['life']]

#model = LinearRegression()
#model.fit(income, life_expectancy)
#print(model)

#Fitting a simple linear regression
linear_reg.fit(income, life_expectancy)

#Printing the solved coefficients for the regression line
print('Life expectancy = m*income + b')
print('Coefficients:', linear_reg.coef_)
print('Intercept:', linear_reg.intercept_)
print('Life expectancy =',linear_reg.intercept_,'+ Income *', linear_reg.coef_)

#Printing the solved coefficients for the regression line
#print('Life expectancy = m*income + b')
#print('Coefficients:', model.coef_)
#print('Intercept:', model.intercept_)
#print('Life expectancy =',linear_reg.intercept_,'+ Income *', linear_reg.coef_)

predicted_life_expectancy = linear_reg.predict(income)
#model.summary()

#Plotting the points and line of fit (model) 
plt.plot(income, life_expectancy, 'bo')
plt.plot(income, predicted_life_expectancy, color='green', linewidth=2)
plt.title("Income vs. Life expectancy and Line of Fit")
plt.xlabel("Income")
plt.ylabel("Life Expectancy")
plt.show()

# Creating the Regression Summary

import statsmodels.api as sm

ydat = raw_data[['life']]
xdat = raw_data[['income']]
xdat = sm.add_constant(xdat)

model = sm.OLS(ydat, xdat)
results = model.fit()
print(results.t_test([1, 0]))
print(results.summary()) 

# The linear equation for life expectancy based on the independent variable 'annual income' is Life expectancy = 38.6717115 + Income * 0.00092173
## The regression summary is also printed below
```

```python
# Analysis 15 : We will now conclude our analysis with a K-means cluster analysis using the whole dataset 
# Cluster Analysis

# Importing the required modules
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans


# Converting catgroy variables to codes
raw_data['Country'] = raw_data['Country'].astype('category')
raw_data['Country'] = raw_data['Country'].cat.codes

raw_data['region'] = raw_data['region'].astype('category')
raw_data['region'] = raw_data['region'].cat.codes

# Building the elbow curve
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(raw_data)
    kmeanModel.fit(raw_data)
    distortions.append(sum(np.min(cdist(raw_data, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / raw_data.shape[0])

# Plotting the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()

# Based on the elbow curve, k=5 seems to be the optimal number of clusters
```

```python
# We will normalize the dataframe before clustering

def Z_ScoreNormalization(data): # we need to normalize the data so that each attribute is brought to same scale

    for i in data.columns:

        mean=data[i].mean()

        std=data[i].std()
        
        data[i]=data[i].apply(lambda d : float(d-mean)/float(std)) #perform z-score normalization
        
Z_ScoreNormalization(raw_data)


# Fitting the kmeans model for k=5

kmeans = KMeans(n_clusters=5, init='k-means++', n_init=10, max_iter=100, tol=0.0001,precompute_distances='auto', verbose=0, random_state=None, copy_x=True, n_jobs=None, algorithm='auto')

kmeans.fit(raw_data)

y_kmeans = kmeans.predict(raw_data)

# Plotting K-means clusters after normalization

plt.scatter(raw_data['income'], raw_data['life'], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=1)
plt.title("K-Means Results")
plt.xlabel("Income")
plt.ylabel("Life Expectancy")
plt.show()


```
